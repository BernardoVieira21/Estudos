```python
# copy, sorted, produtos.sort
# Exercícios
# Aumente os preços dos produtos a seguir em 10%
# Gere novos_produtos por deep copy (cópia profunda)
import copy

produtos = [
    {"nome": "Produto 5", "preco": 10.00},
    {"nome": "Produto 1", "preco": 22.32},
    {"nome": "Produto 3", "preco": 10.11},
    {"nome": "Produto 2", "preco": 105.87},
    {"nome": "Produto 4", "preco": 69.90},
]

produtos = [
    {**produto, "preco": float(f'{produto["preco"] * 1.10:.2f}')}
    for produto in produtos
]

novos_produtos = copy.deepcopy(produtos)

# Ordene os produtos por nome decrescente (do maior para menor)
# Gere produtos_ordenados_por_nome por deep copy (cópia profunda)

def ordenar_nome_decrescente(lista):
    for item in lista:
        print(item)
    print()


produtos_ordenados_por_nome = copy.deepcopy(sorted(novos_produtos, key=lambda item: item["nome"]))

ordenar_nome_decrescente(produtos_ordenados_por_nome)

# Ordene os produtos por preco crescente (do menor para maior)
# Gere produtos_ordenados_por_preco por deep copy (cópia profunda)

def ordenar_preco_crescente(lista):
    for item in lista:
        print(item)
    print()


produtos_ordenados_por_preco = copy.deepcopy(sorted(novos_produtos, key=lambda item: item["preco"]))

ordenar_preco_crescente(produtos_ordenados_por_preco)
```

#### Exercício 8

```python
"""
Considerando duas listas de inteiros ou floats (lista A e lista B)
Some os valores nas listas retornando uma nova lista com os valores somados:
Se uma lista for maior que a outra, a soma só vai considerar o tamanho da
menor.
Exemplo:
lista_a     = [1, 2, 3, 4, 5, 6, 7]
lista_b     = [1, 2, 3, 4]
=================== resultado
lista_soma  = [2, 4, 6, 8]
"""

def zipper(list1, list2):
    interval = min(len(list1), len(list2))
    return [list1[i] + list2[i] for i in range(interval)]


lista_a = [1, 2, 3, 4, 5, 6, 7]
lista_b = [1, 2, 3, 4]

lista_soma = zipper(lista_a, lista_b)
print(lista_soma)
```
